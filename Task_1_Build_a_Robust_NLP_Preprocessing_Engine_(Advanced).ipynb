# TASK 2: Build Advanced Preprocessing Function

```python

def preprocess_text(text):
    """
    Advanced NLP Preprocessing Function
    Steps:
    1. Remove URLs and email patterns
    2. Remove emojis
    3. Convert to lowercase
    4. Handle repeated characters (soooo -> so)
    5. Remove numbers
    6. Remove punctuation
    7. Remove extra spaces
    8. Tokenize
    9. Remove very short tokens (keep 'no', 'not')
    Returns: tokens list, cleaned sentence
    """

    text = re.sub(r'http\S+|www\S+|https\S+', '', text)

    text = re.sub(r'\S+@\S+', '', text)

    text = text.encode('ascii', 'ignore').decode('ascii')

    text = text.lower()

    text = re.sub(r'(.)\1{2,}', r'\1', text)

    text = re.sub(r'\d+', '', text)

    text = text.translate(str.maketrans('', '', string.punctuation))

    text = " ".join(text.split())

    tokens = word_tokenize(text)

    meaningful_short = {'no', 'not'}
    tokens = [w for w in tokens if len(w) > 2 or w in meaningful_short]

    clean_sentence = " ".join(tokens)

    return tokens, clean_sentence

print("Preprocessing function defined!")
```

#Task 4: Token Analytics

```python
print("TOKEN ANALYTICS")


most_noisy_sentence = ""
most_noisy_count = 0
most_meaningful_sentence = ""
most_meaningful_count = 0

for i, sentence in enumerate(test_sentences):
    tokens, clean_sentence = preprocess_text(sentence)

    total_tokens      = len(tokens)
    unique_tokens     = len(set(tokens))
    avg_token_length  = round(sum(len(t) for t in tokens) / total_tokens, 2) if total_tokens > 0 else 0
    noise_count       = len(word_tokenize(sentence)) - total_tokens

    print(f"\nSentence {i+1}: {sentence}")
    print(f"   Total Tokens     : {total_tokens}")
    print(f"   Unique Tokens    : {unique_tokens}")
    print(f"   Avg Token Length : {avg_token_length}")
    print(f"   Noise Removed    : {noise_count} tokens")

    if noise_count > most_noisy_count:
        most_noisy_count = noise_count
        most_noisy_sentence = sentence

    if total_tokens > most_meaningful_count:
        most_meaningful_count = total_tokens
        most_meaningful_sentence = sentence


print("\n" + "=" * 60)
print("ANALYSIS:")
print(f"   Most Noisy Sentence       : {most_noisy_sentence}")
print(f"   Most Meaningful Sentence  : {most_meaningful_sentence}")
print("=" * 60)
```

#Task 5 : Frequency Analysis

```python
from collections import Counter

all_tokens = []
for sentence in test_sentences:
    tokens, _ = preprocess_text(sentence)
    all_tokens.extend(tokens)

word_freq = Counter(all_tokens)

print("FREQUENCY ANALYSIS")

print("\nTop 10 Most Frequent Words:")
for word, count in word_freq.most_common(10):
    print(f"   {word:15} --> {count} times")

print("\nTop 5 Least Frequent Words:")
for word, count in word_freq.most_common()[:-6:-1]:
    print(f"   {word:15} --> {count} times")
```

#Task 6: Build Full Pipeline

```python
def full_pipeline(text_list):
    """
    Full NLP Pipeline
    Input  : List of raw text sentences
    Output : Dictionary with tokens and clean sentences
    """
    all_tokens        = []
    all_clean_sentences = []

    for text in text_list:
        tokens, clean_sentence = preprocess_text(text)
        all_tokens.append(tokens)
        all_clean_sentences.append(clean_sentence)

    return {
        "tokens": all_tokens,
        "clean_sentences": all_clean_sentences
    }

result = full_pipeline(test_sentences)

print("FULL PIPELINE OUTPUT")

for i, (tokens, sentence) in enumerate(zip(result["tokens"], result["clean_sentences"])):
    print(f"\nSentence {i+1}:")
    print(f"   Tokens         : {tokens}")
    print(f"   Clean Sentence : {sentence}")

print("\n Full Pipeline executed successfully!")
```
